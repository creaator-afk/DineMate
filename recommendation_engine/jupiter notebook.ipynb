# Create vectorization of "tags" column

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(max_features=6000, stop_words="english")
```

```python
vectors = cv.fit_transform(new_df["tags"]).toarray()
```

```python
vectors.shape
```

```python
count = 0
for i in cv.get_feature_names_out():
    print(i)
    count+=1
print(count)
```

# Apply stemming

```python
from nltk.stem.porter import PorterStemmer

ps = PorterStemmer()
```

```python
ps.stem("lovely")
```

```python
def stemming_now(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_df["tags"] = new_df["tags"].apply(stemming_now)
```

```python
# now after applying stemming
new_df.iloc[0,2]
```

```python
# before apply stemming
df.iloc[0,-1]
```

# Calculate similarity of vectors

```python
from sklearn.metrics.pairwise import cosine_similarity

similarity = cosine_similarity(vectors)
```

```python
similarity.shape
```

# Food Recommendation

```python
def recommend_food(food_name):
    food_name_index = new_df[new_df["name"]==food_name].index[0]
    print(food_name_index)
    distances = similarity[food_name_index]
    recommended_foods = sorted(list(enumerate(distances)), reverse=True, key=lambda x:x[1])[1:7]
    current = [new_df.iloc[food_name_index,0], new_df.iloc[food_name_index,1]]
    for i in recommended_foods:
        index = i[0]
        print(new_df.iloc[index][0])
    print(current)
```

```python
recommend_food("Spicy Cabbage Rice  (South Indian Style Mutta Kose Sadham )")
```
